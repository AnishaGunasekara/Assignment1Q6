```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
im_og = cv.imread('/content/drive/MyDrive/Image_Processing/einstein.png')
im_og = cv.cvtColor(im_og, cv.COLOR_BGR2GRAY)

im_1 = cv.GaussianBlur(im_og, (3,3),0)

sobelx = cv.Sobel(im_1, cv.CV_64F, 1,0, ksize=5)
sobely = cv.Sobel(im_1, cv.CV_64F, 0,1, ksize=5)


fig,ax = plt.subplots(1,3,figsize =(10,10))
ax[0].imshow(im_og, cmap='gray')
ax[0].set_title('Original Image')
ax[1].imshow(sobelx, cmap='gray')
ax[1].set_title('Sobel-x Edge Detection')
ax[2].imshow(sobely, cmap='gray')
ax[2].set_title('Sobel-y Edge Detection')

plt.show()


```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
im_og = cv.imread('/content/drive/MyDrive/Image_Processing/einstein.png')
im_og = cv.cvtColor(im_og, cv.COLOR_BGR2GRAY)

def sobel_filter(image):

    kernel_x = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
    kernel_y = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])

    gradient_x = cv.filter2D(im_og, cv.CV_64F, kernel_x)
    gradient_y = cv.filter2D(im_og, cv.CV_64F, kernel_y)

    magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

    return gradient_x, gradient_y, magnitude

gradient_x, gradient_y, magnitude = sobel_filter(im_og)

plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1), plt.imshow(im_og, cmap='gray')
plt.title('Original Image'), plt.axis('off')

plt.subplot(2, 2, 2), plt.imshow(gradient_x, cmap='gray')
plt.title('Sobel X'), plt.axis('off')

plt.subplot(2, 2, 3), plt.imshow(gradient_y, cmap='gray')
plt.title('Sobel Y'), plt.axis('off')

plt.subplot(2, 2, 4), plt.imshow(magnitude, cmap='gray')
plt.title('Magnitude'), plt.axis('off')

plt.show()

```
